# Importing necessary libraries

```python
import numpy as np  # NumPy is used for numerical operations and working with arrays
import pandas as pd  # Pandas is used for data manipulation and analysis, particularly for reading the dataset
import tensorflow as tf  # TensorFlow is a library for machine learning and deep learning models
from sklearn.preprocessing import MinMaxScaler  # MinMaxScaler is used to scale features to a given range (here, [0, 1])
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  # Metrics to evaluate model performance
import matplotlib.pyplot as plt  # Matplotlib is used for plotting graphs
from tensorflow.keras.models import Sequential  # Sequential model allows us to build layers step by step
from tensorflow.keras.layers import Dense  # Dense layer is a fully connected layer (each neuron connected to all inputs)
```

# Dataset loading

```python
data = pd.read_csv("D:/noody/Deep learning/Lab 1/orbit.csv")  # Reading the CSV file containing the orbital data
# 'data' is now a DataFrame with columns 'time_steps' and 'y' (positions)
```

# Missing values check

```python
print("Missing values before handling:\n", data.isnull().sum())  
# 'isnull().sum()' checks for any missing values (NaN) in each column and prints the total count of NaNs for each column

# Handle missing values by filling with the column mean
data['time_steps'].fillna(data['time_steps'].mean(), inplace=True)  
# Fills missing values in the 'time_steps' column with the mean of that column
data['y'].fillna(data['y'].mean(), inplace=True)  
# Fills missing values in the 'y' (positions) column with the mean of that column

# Print missing values after filling them (to confirm there are no more NaNs)
print("Missing values after handling:\n", data.isnull().sum())  
# Verifies that there are no more missing values in the dataset
```

# Data normalization

```python
scaler = MinMaxScaler()  # Initialize a MinMaxScaler to scale the data to a range of [0, 1]
time_steps = scaler.fit_transform(data['time_steps'].values.reshape(-1, 1))  
# Reshape time_steps from a 1D array to a 2D array and scale it to [0, 1] using fit_transform()
# Reshaping is needed because the model expects inputs in 2D format (batch_size, feature_size)
y_positions = scaler.fit_transform(data['y'].values.reshape(-1, 1))  
# Similarly, scale the 'y' positions to the [0, 1] range
```

# Sequential neural network model

```python
model = Sequential()  # Instantiate a sequential model, where layers are added one by one
# Adding the first layer with 128 neurons and 'relu' activation function
model.add(Dense(128, input_shape=(1, ), activation='relu'))  
# 'input_shape=(1,)' because we are passing 1D time_steps, 'relu' activation helps capture non-linearity

# Add hidden layers with decreasing number of neurons to gradually reduce complexity
model.add(Dense(64, activation='relu'))  # 64 neurons, 'relu' activation
model.add(Dense(32, activation='relu'))  # 32 neurons, 'relu' activation
model.add(Dense(16, activation='relu'))  # 16 neurons, 'relu' activation

model.add(Dense(1)) # Output layer with 1 neuron (since we are predicting a single value: the position 'y')

model.summary()  # Print a summary of the model architecture (layers and number of parameters)
```

# Compiling and training the model

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])  # Adam optimizer, MSE as loss, and MAE as an additional evaluation metric

history = model.fit(time_steps, y_positions, epochs=100, validation_split=0.2)  # Train the model for 100 epochs, with 20% of data used for validation

print('Training completed.')  # Indicate that training has finished
```

# Evaluating the model

```python
predictions = model.predict(time_steps)  # Use the model to predict the 'y' values (positions) for the given 'time_steps'

# Rescale predictions and actual values back to their original (non-normalized) scale
y_positions_original = scaler.inverse_transform(y_positions)  # Rescale the actual 'y' values to their original scale
predictions_original = scaler.inverse_transform(predictions)  # Rescale the predicted 'y' values to their original scale
```

# Calculate performance metrics

```python
mse = mean_squared_error(y_positions_original, predictions_original)  # Compute the Mean Squared Error (MSE)
mae = mean_absolute_error(y_positions_original, predictions_original)  # Compute the Mean Absolute Error (MAE)
r2 = r2_score(y_positions_original, predictions_original)  # Compute the R-squared (R²) score, which measures how well the model fits the data

# Print the performance metrics
print(f'MSE: {mse}')  # Print the MSE value
print(f'MAE: {mae}')  # Print the MAE value
print(f'R-squared: {r2}')  # Print the R² value
```

# Plot actual vs predicted positions to visualize model performance

```python
plt.figure(figsize=(10, 6))  # Create a new figure with size 10x6 inches
plt.plot(data['time_steps'], y_positions_original, label='Actual Orbit', color="mediumslateblue")  # Plot actual positions (in blue)
plt.plot(data['time_steps'], predictions_original, label='Predicted Orbit', color="orange")  # Plot predicted positions (in orange)
plt.legend(loc="lower left")  # Place the legend in the lower-left corner
plt.title(f"Predicted Orbit vs Actual Orbit (R² = {r2:.2f})")  # Title of the plot, including the R² value
plt.xlabel('Time Steps')  # Label for the X-axis (time steps)
plt.ylabel('Position')  # Label for the Y-axis (position)
plt.show()  # Display the plot
```

```python
# Scatter plot with a line connecting the points
plt.figure(figsize=(10, 6))  # Create a new figure with size 10x6 inches
plt.scatter(data['time_steps'], y_positions_original, label='Actual Orbit', color="blue")  # Scatter plot of actual positions (in blue)
plt.plot(data['time_steps'], predictions_original, label='Predicted Orbit', color="red", linestyle='--')  # Dashed line plot for predictions (in red)
plt.legend(loc="upper left")  # Place the legend in the upper-left corner
plt.title(f"Scatter Plot: Predicted vs Actual Positions (R² = {r2:.2f})")  # Title for the alternative visualization
plt.xlabel('Time Steps')  # X-axis label
plt.ylabel('Position')  # Y-axis label
plt.show()  # Display the scatter plot
```
